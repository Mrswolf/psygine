## GoPro Blurring Dataset

```python
from psygine.datasets.image import GoProTrainDataset, GoProTestDataset
import matplotlib.pyplot as plt

dataset = GoProTrainDataset(local_path=local_path, prefetch=False)

blur, sharp = dataset[0]

fig, axs = plt.subplots(1, 2)
axs[0].imshow(blur)
axs[0].axis('off')
axs[0].title.set_text('Blur')
axs[1].imshow(sharp)
axs[1].axis('off') 
axs[1].title.set_text('Sharp')
fig.tight_layout()
```

## YaleFace Dataset

```python
from psygine.datasets.image import YaleFaceBDataset
import matplotlib.pyplot as plt

dataset = YaleFaceBDataset(local_path=local_path)

faces = dataset[0]

fig, axs = plt.subplots(1, 4)
for i in range(4):
    axs[i].imshow(faces[i], cmap='gray')
    axs[i].axis('off')
    axs[i].title.set_text(f'Face {i+1}')
```

## Misc

```python
from psygine.datasets.misc import TrafficVideo1Dataset
import matplotlib.pyplot as plt

dataset = TrafficVideo1Dataset(local_path=local_path)
frames = dataset[0]

fig, axs = plt.subplots(1, 4)
for i in range(4):
    axs[i].imshow(frames[i * frames.shape[0] // 4], cmap="gray")
    axs[i].axis("off")
    axs[i].title.set_text(f"Frame {i * frames.shape[0] // 4 + 1}")
```
